# PV Curve Shape Analysis

Load one Home Assistant history export, estimate PV power from one base elevation formula plus azimuth shape points, and compare measured and estimated power in one chart.

The shape multiplier defaults to `1.0` everywhere. User-defined shape points are connected linearly, with implicit endpoints at azimuth `0` and `360` set to `1.0`.

```python
import importlib
from pathlib import Path

import analyze_pv_curve
import plotly.io as pio

analyze_pv_curve = importlib.reload(analyze_pv_curve)
EstimateParameters = analyze_pv_curve.EstimateParameters
HaEntityNames = analyze_pv_curve.HaEntityNames
add_estimate_column = analyze_pv_curve.add_estimate_column
build_figure = analyze_pv_curve.build_figure
build_shape_figure = analyze_pv_curve.build_shape_figure
load_input_csv = analyze_pv_curve.load_input_csv
normalize_shape_points = analyze_pv_curve.normalize_shape_points

pio.renderers.default = "vscode"

csv_path = Path("data/history.csv")
entity_names = HaEntityNames(
    measured_power_sensor="sensor.fems_productiondcactualpower",
    azimuth_sensor="sensor.smart_cover_automation_sonnenazimut",
    elevation_sensor="sensor.smart_cover_automation_sonnenhohe",
)

shape_points = normalize_shape_points(
    (
        # Example azimuth/multiplier points:
        # (170.0, 0.9),
        # (210.0, 0.75),
    )
)

params = EstimateParameters(
    peak_power_w=6800.0,
    min_elevation_deg=0,
    exponent=0.8,
    shape_points=shape_points,
)
```

```python
data = load_input_csv(csv_path, entity_names=entity_names)
analyzed = add_estimate_column(data, params)
figure = build_figure(analyzed, "Measured vs estimated PV power", params)
figure
```

```python
shape_figure = build_shape_figure(analyzed, params)
shape_figure
```
